## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    freq = {}

    for item in arr:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    mode_value = 0
    max_count = 0

    for item in freq:
        if freq[item] > max_count:
            max_count = freq[item]
            mode_value = item

    return mode_value
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def standard_deviation(data):
    # Step 1: Calculate the mean manually
    total = 0
    count = 0
    for x in data:
        total += x
        count += 1
    mean = total / count

    # Step 2: Calculate variance manually
    squared_diff_sum = 0
    for x in data:
        squared_diff_sum += (x - mean) ** 2
    variance = squared_diff_sum / (count - 1)

    # Step 3: Square root using exponentiation
    std_dev = variance ** 0.5
    return std_dev

```

```python
def is_pangram(s):
    # Convert string to lowercase
    s = s.lower()

    # Iterate through all letters in the alphabet
    for char in 'abcdefghijklmnopqrstuvwxyz':
        if char not in s:
            return False  # Missing a letter
    return True  # All letters found

# Examples
#print(is_pangram("The quick brown fox jumps over the lazy dog"))  # True
#print(is_pangram("Hello World!"))  # False
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(s):
    words = []
    word = ""

    # Step 1: manually split by comma
    for char in s:
        if char == ',':
            words.append(word)  # end of word
            word = ""
        else:
            word += char
    words.append(word)  # add the last word

    # Step 2: sort the list
    sorted_words = sorted(words)

    # Step 3: manually create comma-separated string
    result = ""
    for i, w in enumerate(sorted_words):
        result += w
        if i != len(sorted_words) - 1:
            result += ","
    return result

# Example
#s = "banana,apple,orange,grape"
#print(sort_csv_words(s))
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    spec_char = "!@#$%^&*()[]^'{-}+?_=,<>/"
    if (len(password) >= 8) and (any(c.isupper() for c in password)) and (any(c.islower() for c in password)) and (any(c.isdigit() for c in password)) and (any(c in spec_char for c in password)):
        return True
    else:
        return False
```
